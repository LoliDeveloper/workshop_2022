```python
import lasio
import os
from os import path

dir = r'D:\MyFolder\1.Education\2.Self-Ed\БММ 2022\dataset\РИГИС'

holder = []
for filename in os.listdir(dir):
    if '.las' in filename:
        _p = path.join(dir, filename)
        las = lasio.read(_p)
        wname = filename[:-4]
        df = las.df()
        df['wname'] = wname
        holder.append(df)
        
```

```python
import pandas as pd
df = pd.concat(holder, axis = 0)

df1 = df[df['LIT_T']==1]
df1['KP_T'].plot(kind = 'kde')
```
